```python
from scipy import signal
import numpy as np
a_low = [1, -2, 1]
b_low = np.concatenate(([1], np.zeros(5), [-2], np.zeros(5), [1]))

w, h = signal.freqz(b_low, a_low)
f = w * 200 / (2*np.pi)
import matplotlib.pyplot as plt
fig = plt.figure()
plt.title('Digital filter frequency response')
ax1 = fig.add_subplot(111)


plt.plot(f, 20 * np.log10(abs(h)), 'b')
plt.ylabel('Amplitude [dB]', color='b')
#plt.xlabel('Frequency [rad/sample]')
plt.xlabel('Frequency [hz]')
plt.show()
```

```python
from scipy import signal
import numpy as np
a_high = [1,-1]
b_high = np.concatenate(([-1/32], np.zeros(15), [1, -1], np.zeros(14), [1/32]))

w, h = signal.freqz(b_high, a_high)
f = w * 200 / (2*np.pi)
import matplotlib.pyplot as plt
fig = plt.figure()
plt.title('Digital filter frequency response')
ax1 = fig.add_subplot(111)


plt.plot(f, 20 * np.log10(abs(h)), 'b')
plt.ylabel('Amplitude [dB]', color='b')
#plt.xlabel('Frequency [rad/sample]')
plt.xlabel('Frequency [hz]')
plt.show()
```

```python
import wfdb
from scipy.signal import lfilter

record = wfdb.rdsamp('100',pbdir='mitdb', sampto = 2000, channels=[0])

siglow = lfilter(b_low, a_low, record.p_signals)

sighigh = lfilter(b_high, a_high, record.p_signals)

plt.plot(siglow)
plt.show()

plt.plot(sighigh)
plt.show()

sigband = lfilter(b_high, a_high, siglow)

plt.plot(sigband)
plt.show()
```

```python
from scipy import signal
import matplotlib.pyplot as plt


b, a = signal.butter(2, np.array([5, 15])/100, 'bandpass')

sigpass = signal.lfilter(b, a, record.p_signals)

plt.plot(sigpass)
plt.show()
```
