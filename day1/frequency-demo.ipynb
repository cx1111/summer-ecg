```python
import math

import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.fftpack import fft
```

```python
t = np.linspace(0, 6*math.pi)
x = np.sin(t)
x2 = -x
```

```python
line_x, = plt.plot(x, label='x')
line_x2, = plt.plot(x2, label='x1')
plt.legend(handles=[line_x, line_x2])
```

```python
fs = 50 / max(t)
N = len(t)

f, Pxx_den = signal.periodogram(x, fs)
f, Pxx_den2 = signal.periodogram(x2, fs)

line_pxx = plt.plot(f, Pxx_den)[0]
line_pxx2 = plt.plot(f, Pxx_den2)[0]

#plt.ylim([1e-7, 1e2])
plt.legend(handles=[line_pxx, line_pxx2])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()


```

```python
y = fft(x)
y2 = fft(x2)
```

```python
plt.plot(y.real[:25])
plt.plot(y2.real[:25])
plt.title('Magnitude')
```

```python
plt.plot(y.imag[:25])
plt.plot(y2.imag[:25])
plt.title('Phase')
```

```python
plt.plot(np.abs(y[:25]))
plt.plot(np.abs(y2[:25]))
plt.title('Amplitude')
```
